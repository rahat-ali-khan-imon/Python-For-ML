# Operators in Python

## Arithmetic Operators

```python
x = 10; y = 3

sum = x + y     # Summation
sub = x - y     # Subtraction
mult = x * y    # Multiplication
div = x / y     # Division
round_number = round(x / y)
div2 = round(x / y, 5)
rem = x % y     # Remainder
integer_number = x // y
power = x ** y

print(sum, sub, mult, div, div2, rem)
print("Summation: ", sum)
print("Subtraction: ", sub)
print("Multiplication: ", mult)
print("Division: ", div)
print("Round Number: ", round_number)
print("Division and round: ", div2)
print("Remainder: ", rem)
print("Integer Number: ", integer_number)
print("Exponentiation/Power: ", power)
```

```python
x, y = 10, 3

sum = x + y     # Summation
sub = x - y     # Subtraction
mult = x * y    # Multiplication
div = x / y     # Division
round_number = round(x / y)
div2 = round(x / y, 5)
rem = x % y     # Remainder
integer_number = x // y
power = x ** y

print(sum, sub, mult, div, div2, rem)
print(f"Summation {x} and {y}: ", sum)
print(f"Subtraction {x} and {y}: ", sub)
print(f"Multiplication {x} and {y}: ", mult)
print(f"Division {x} and {y}: ", div)
print(f"Round Number {x} and {y}: ", round_number)
print(f"Division and round {x} and {y}: ", div2)
print(f"Remainder {x} and {y}: ", rem)
print(f"Integer Number {x} and {y}: ", integer_number)
print(f"Exponentiation/Power {x} and {y}: ", power)
```

```python
# Basic Arithmetic Operations with Two Numbers

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent (Power):", a ** b)

# 🧠 Explanation:
# + → যোগ (Addition)
# - → বিয়োগ (Subtraction)
# * → গুণ (Multiplication)
# / → ভাগ (Normal Division — gives float result)
# // → Floor Division (quotient without decimal part)
# % → Modulus (remainder after division)
# ** → Exponentiation (power)
```

## Comparison Operators

```python
# >, >=, <, <=, ==, !=

x = 10
y = 3

greater_than = x > y
greater_than_equal = (10 >= 10)

print(greater_than, greater_than)


less_than = x < y
less_than_equal = (10 <= 10)

print(less_than, less_than_equal)


equal1 = x == y
equal2 = x == 10

print(equal1, equal2)


not_equal1 = x != y
not_equal2 = x != 10

print(not_equal1, not_equal2)


print(x == y, x > y, x != y)
```

## Membership Operator

```python
print("a" in "apple")
print("b" not in "apple")
print("A" in "apple")
print("le" in "apple")
print("\n")

string = "Hi, Hello, I am robot and You ?"

print("Hi" in string)
print("Hi" not in string)
print("Robot" in string)
print("robot" not in string)
print("robot" in string)
print("Hey" not in string)
print("I am robot" in string)
```

```python
# Membership Operator with String

text = "Python Programming"

print('P' in text)        # True
print('python' in text)   # False (case-sensitive)
print('thon' in text)     # True
print('z' not in text)    # True

# 🧠 What Are Membership Operators?
# 👉 Membership operators দিয়ে চেক করা হয়
# কোনো element কোনো sequence (string, list, tuple, set, dictionary) এর মধ্যে আছে কিনা।

# 🔸 Two Membership Operators
# | Operator | Meaning                                  | Example              | Result |
# | -------- | ---------------------------------------- | -------------------- | ------ |
# | `in`     | Returns True if value is present         | `'a' in 'apple'`     | True   |
# | `not in` | Returns True if value is **not** present | `'x' not in 'apple'` | True   |
```

```python
# Membership Operator with List

nums = [10, 20, 30, 40, 50]

print(20 in nums)       # True
print(60 not in nums)   # True
print(10 not in nums)   # False
```

```python
# Membership Operator with Tuple

colors = ('red', 'green', 'blue')

print('green' in colors)      # True
print('yellow' not in colors) # True
```

```python
# Membership Operator with Set

fruits = {'apple', 'banana', 'cherry'}

print('apple' in fruits)       # True
print('grape' not in fruits)   # True

# Set membership checking is very fast — Python uses hashing for it
```

```python
# Membership Operator with Dictionary

student = {'name': 'Alice', 'age': 21, 'grade': 'A'}

print('name' in student)       # True (key)
print('Alice' in student)      # False (value)
print('grade' not in student)  # False


# 🧠 Important:
# Membership check in a dictionary checks keys only, not values.
```

```python
# Membership Operator with Condition

ch = input("Enter a letter:").lower()
vowels = "aeiou"

if ch in vowels:
    print("Vowel")
else:
    print("Consonant")

# Classic real-life use case of in.
```

```python
# Membership Operator with Logical Combination

word = "data"

if 'a' in word and 't' in word:
    print("Both letters are present")
else:
    print("At least one letter missing")
```

```python
# Membership Operator in Nested Structures

records = [[1, 'Alice'], [2, 'Bob'], [3, 'Charlie']]

print([2, 'Bob'] in records)    # True
print(['Bob', 2] in records)    # False (order matters)
```

```python
# Check if a Number Exists in a List

# Take a list of numbers
numbers = [10, 20, 30, 40, 50]

# Take input from user
num = int(input("Enter a number: "))

# Check membership
if num in numbers:
    print(num, "is in the list")
else:
    print(num, "is not in the list")
```

```python
# Check if a Character Exists in a String

word = input("Enter a word: ")
ch = input("Enter a character to check: ")

if ch in word:
    print(f"'{ch}' is found in '{word}'")
else:
    print(f"'{ch}' is NOT found in '{word}'")
```

```python
# Check if an Item Exists in a Tuple

fruits = ("apple", "banana", "mango")
item = input("Enter a fruit name: ")

if item in fruits:
    print(item, "is available")
else:
    print(item, "is not available")
```

```python
# Using not in

sentence = input("Enter a sentence: ")
word = input("Enter a word to check: ")

if word not in sentence:
    print(f"'{word}' is NOT present in the sentence.")
else:
    print(f"'{word}' is present in the sentence.")
```

## Identity Operators

```python
x, y = 10, 20

print(x is y)
print(x is not y)

ans = x is y
result = x is not y

print(ans)
print(result)
```

```python
# Basic Example

# 🧠 What Are Identity Operators?
# Identity operators check whether two variables point to the same object in memory, not just whether they have equal values.
# 👉 অর্থাৎ, এটা memory location compare করে — value না।
# এই অপারেটরগুলো দিয়ে আমরা বুঝি দুইটা ভ্যারিয়েবল একই মেমরি লোকেশন নির্দেশ করছে কি না।

x = 10
y = 10

print(x == y)   # True (values are same)
print(x is y)   # True (both refer to same memory location)

# Python internally stores small integers (-5 to 256) in the same memory, so x is y → True here.

# 🔸 Two Identity Operators
# | Operator | Meaning                                             | Example      | Output       |
# | -------- | --------------------------------------------------- | ------------ | ------------ |
# | `is`     | True if both variables refer to the **same object** | `a is b`     | True / False |
# | `is not` | True if they refer to **different objects**         | `a is not b` | True / False |
```

```python
# When “is” and “==” Give Different Results

a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   # True (same values)
print(a is b)   # False (different memory locations)


# 🧠 Explanation:
# == checks values
# is checks memory identity
# Even though both lists look identical, they are stored separately in memory.
```

```python
# Example with is not

x = [10, 20]
y = [10, 20]

print(x is not y)  # True (different memory objects)
print(x != y)      # False (values are same)
```

```python
# Using is with None (Most Common Use)

result = None

if result is None:
    print("No result found!")
else:
    print("Result available!")

# 🧠 Always use is / is not with None instead of ==.
```

```python
# Example with Strings (Caching)

a = "hello"
b = "hello"

print(a == b)   # True
print(a is b)   # True (Python caches short strings)

# But 👇

x = "hello world!"
y = "hello world!"

print(x == y)   # True
print(x is y)   # Might be False (depends on memory optimization)
```

```python
# Example with Booleans

x = True
y = True

print(x is y)   # True
print(x == y)   # True

# Because both point to the same singleton Boolean objects in memory.
```

```python
# Example with Tuples (Immutable)

t1 = (1, 2, 3)
t2 = (1, 2, 3)

print(t1 == t2)  # True
print(t1 is t2)  # False (separate objects)
```

```python
# ⚠️ Common Interview Trick

x = [1, 2, 3]
y = x

print(x == y)  # True
print(x is y)  # True

y = [1, 2, 3]
print(x == y)  # True
print(x is y)  # False

# 💬 In short:
# == → compares value/content
# is → compares memory identity (address)
```

```python
# 🧩 Concept in Simple Words

# 👉 == checks values (দুইটার মান সমান কি না)
# 👉 is checks identity (দুইটা ভ্যারিয়েবল মেমরিতে একই object কে নির্দেশ করছে কি না)

# Simple Identity Check

# Take user input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("a == b:", a == b)
print("a is b:", a is b)
print("a is not b:", a is not b)

# 🧠 Explanation:
# a == b → checks if values are equal
# a is b → checks if both variables refer to the same memory object
# a is not b → opposite of is

# 🧩 Note: ছোট integer (–5 থেকে 256 পর্যন্ত) Python মেমরিতে cache করে রাখে, তাই অনেক সময় is ও True দেয়।
```

```python
# Lists Comparison (Identity vs Equality)

# Take two lists from user
list1 = list(map(int, input("Enter list 1 elements: ").split()))
list2 = list(map(int, input("Enter list 2 elements: ").split()))

print("list1 == list2:", list1 == list2)
print("list1 is list2:", list1 is list2)

# 💡 কারণ দুইটা list আলাদা জায়গায় তৈরি হয়েছে, তাই তাদের identity আলাদা।
```

```python
# Assigning Reference

x = [10, 20, 30]
y = x   # same object reference

print("x == y:", x == y)
print("x is y:", x is y)

# 🧠 এখানে y = x মানে y এবং x দুইটাই একই list কে নির্দেশ করছে।
```
