## C. Even, Odd, Positive and Negative

```python
'''https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/C'''
```

```python
n = int(input())

numbers = list(map(int, input().split()))

even = odd = positive = negative = 0

for i in numbers:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
    
    if i > 0:
        positive += 1
    else:   # elif i < 0:
        if i == 0:
            continue
        negative += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

## Q. Digits

```python
'''https://codeforces.com/group/MWSDmqGsZm/contest/219432/problem/Q'''
```

```python
t = int(input())

for i in range(t):
    n = int(input())

    if n == 0:
        print(0)
        continue

    while n != 0:
        digit = n % 10
        n //= 10
        print(digit, end = " ")
    print()
```

```python
t = int(input())
 
for i in range(t):
    n = int(input())
    if n == 0:
        print(0)
        continue
    while n > 0:
        print(n % 10, end = ' ')
        n //= 10
    print()

# Input:
# 121
# 39
# 123456
# 1200
```
