# Break and Continue Statements in Python

## Break Statement

```python
# Basic break Example

for i in range(1, 6):
    if i == 4:
        break
    print(i)
print("Loop ended")

# break → loop পুরোপুরি থেমে যাবে (it immediately exits the loop).
```

```python
# break inside while loop

i = 1
while i <= 5:
    if i == 4:
        break
    print(i)
    i += 1
print("Loop stopped at i =", i)
```

```python
# break in nested loop

for i in range(1, 4):
    for j in range(1, 4):
        if j == 2:
            break
        print(f"i = {i}, j = {j}")

# 🧠 Explanation:
# Outer loop → i 1 থেকে 3 পর্যন্ত।
# Inner loop → j 1 থেকে 3 পর্যন্ত।
# যখন j == 2, তখন inner loop থেমে যায় (outer loop চালু থাকে)।
```

## Continue Statement

```python
# Basic continue Example

for i in range(1, 6):
    if i == 3:
        continue
    print(i)
    
# 🧠 Explanation:
# continue → current iteration skip করে পরের iteration এ চলে যায়।
# যখন i == 3, তখন print skip হবে।
# বাকিগুলো print হবে।
```

```python
# continue inside while loop

i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
# continue in nested loop

for i in range(1, 4):
    for j in range(1, 4):
        if j == 2:
            continue
        print(f"i={i}, j={j}")

# 🧠 Explanation:
# এখানে continue শুধু inner loop এর current iteration skip করছে।
# j == 2 হলে skip, বাকিগুলো print হবে।
```

```python
# break and continue together

for i in range(1, 6):
    if i == 2:
        continue  # Skip 2
    if i == 5:
        break     # Stop at 5
    print(i)

# 🧠 Explanation:
# যখন i == 2, তখন continue → skip print
# যখন i == 5, তখন break → loop stop
# তাই print হবে শুধুমাত্র 1, 3, 4
```

```python
# Practical Example — Skip negative numbers

numbers = [5, -3, 8, -2, 7]

for n in numbers:
    if n < 0:
        continue
    print(n)
```
