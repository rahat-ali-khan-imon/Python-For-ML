# Variables and Data Types in Python

## Variables and Data Types

```python
phone_number = 1234567

print(phone_number)
print(type(phone_number))
```

```python
name = "Phitron"

print(name)
print(type(name))
```

```python
age = 23
height = 5.7
name = "Alice"
is_passed = True

print(type(age), type(height), type(name), type(is_passed))

age = 23.5
print(age, type(age))
```

```python
# Variable Basic Example

name = "Alice"
age = 25
height = 5.6
is_student = True

print(name, type(name))
print(age, type(age))
print(height, type(height))
print(is_student, type(is_student))
```

```python
# Multiple Assignment

x, y, z = 10, 20, 30

print(x, y, z)
print(type(x))
```

```python
# Single Value to Multiple Variables

a = b = c = 5

print(a, b, c)
print(type(a))
```

```python
# Type Checking

x = 10
y = 10.5
z = "Python"
w = True

print(type(x))
print(type(y))
print(type(z))
print(type(w))
```

```python
# Type Casting (Changing Data Type)

x = 10
y = float(x)
z = str(x)

print(y, type(y))   # 10.0
print(z, type(z))   # "10"
```

```python
# Input and Type Conversion

name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Hello", name, "! You are", age, "years old.")
```

```python
# Dynamic Typing
# Python automatically changes variable type:

x = 10
print(x, type(x))

x = "Python"
print(x, type(x))
```

```python
# Complex Number Example

c = 3 + 4j

print(c.real)
print(c.imag)
print(type(c))
```

```python
# Boolean Example

x = 10
y = 20

print(x > y)  # False
print(x < y)  # True
```

```python
# Type Conversion with Boolean

print(bool(0))      # False
print(bool(""))     # False
print(bool(" "))    # True
print(bool(1))      # True
print(bool("Hi"))   # True
print(bool(5))      # True
```

```python
# Example of None Type

x = None

print(x)
print(type(x))
```

```python
# Changing Variable Type Example

a = "100"

b = int(a) + 50
print(b)
```

```python
# Checking Data Type using isinstance()

x = 5.5

print(isinstance(x, float))   # True
print(isinstance(x, int))     # False

# 🧠 Explanation:
# isinstance(x, float)
# → এই function টি চেক করে যে x কি float টাইপের কিনা।
# যেহেতু x = 5.5, তাই এটা float, তাই output হবে True.

# isinstance(x, int)
# → এটা চেক করছে x কি integer কিনা।
# যেহেতু x আসলে float, তাই এই condition False হবে।
```

## User Input

```python
# Integer Input

num = int(input("Enter an integer: "))

print("You entered:", num)
print("Data type:", type(num))

# 🧠 Explanation:
# input() always returns a string by default।
# int() দিয়ে string → integer convert করা হয়।
```

```python
# Float Input

price = float(input("Enter price: "))

print("Price is:", price)
print("Data type:", type(price))

# 🧠 Explanation:
# float() → string → floating-point number।
```

```python
# String Input

name = input("Enter your name: ")

print("Hello,", name)
print("Data type:", type(name))

# 🧠 Explanation:
# input() string return করে।
# কোনো conversion না করলে data type always <class 'str'>।
```

```python
# Boolean Input

flag = input("Enter True/False: ").lower()=="true"

print("Boolean value:", flag)
print("Data type:", type(flag))

# 🧠 Explanation:
# User input string → lowercase → compare "true"
# True/False boolean value তৈরি হয়।
# Example input: True
```

```python
# Multiple Values Input (List of Integers)

numbers = list(map(int, input("Enter numbers separated by space: ").split()))

print("Numbers:", numbers)
print("Data type:", type(numbers), type(numbers[0]))

# 🧠 Explanation:
# input() → string → .split() → list of strings
# map(int, ...) → strings → integers
# list() → map object → list
# Example input: 1 2 3 4 5
```

```python
n = input()

print(n)
print(type(n), type(n[0]))
```

```python
n = input().split()

print(n)
print(type(n), type(n[0])) 
```

```python
n = map(int, input().split())

print(n)
print(type(n))
```
