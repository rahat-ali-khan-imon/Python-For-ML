# For and While Loop in Python

## Foor Loop

```python
# For Loop Syntax ---> for variable in iterable:

for i in range(10):  # range(stop) ---> 0 to n - 1
    print(i, "Hello Python")

list(range(10))
```

```python
for i in range(11):
    print(i, "Hello Python")

print(list(range(11)))
```

```python
for i in range(1, 11):  # range(start, stop)
    print(i, "Hello Python")

list(range(1, 11))
```

```python
for _ in range(1, 11):  # range(start, stop)
    print("Hello Python")

list(range(1, 11))
```

```python
for i in range(1, 11, 2):  # range(start, stop, step)
    print(i, "Hello Python")

list(range(1, 11, 2))
```

```python
# Basic for Loop

for i in range(5):
    print(i)

# 🧠 এখানে range(5) মানে 0 থেকে 4 পর্যন্ত সংখ্যা — মোট 5 বার loop হবে।
```

```python
for i in range(6):
    print(i, end=" ")
```

```python
# Loop with Start and End

for i in range(2, 7):
    print(i)

# 🧠 এখানে range(2, 7) মানে শুরু 2 থেকে, শেষের আগ পর্যন্ত (7 বাদে)।
```

```python
# Loop with Step Value

for i in range(0, 10, 2):
    print(i)

# 🧠 এখানে range(start, end, step) — মানে প্রতি ধাপে 2 করে বাড়ছে।
```

```python
# Reverse Loop

for i in range(10, 0, -1):  # range(start, end, step)
    print(i, end=" ")
print()

print(list(range(10, 0, -1)))

# 🧠 এখানে -1 মানে — প্রতি ধাপে 1 করে কমবে।
```

```python
# Loop Over a List

fruits = ["apple", "banana", "cherry"]
for f in fruits:
    print(f)
```

```python
# Loop Over a String

for ch in "Python":
    print(ch)
```

```python
string = "Hi, Hello, Python"

for ch in string:
    print(ch)

print(list(string))
```

```python
# Loop with Condition

for i in range(1, 6):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
```

```python
# Using else with for loop

for i in range(3):
    print(i)
else:
    print("Loop finished successfully!")

# 🧠 else part তখনই চলবে যদি loop break না হয়।
```

```python
for i in range(3):
    print(i)
    if i == 2:
        break
else:
    print("Loop finished successfully!")
```

```python
# Right method

for i in range(3): 
    if i == 2:
        break;
    print(i)
else:
    print("Loop finished successfully!")
```

```python
for i in range(3):
    print(i)
    if i == 2:
        continue
else:
    print("Loop finished successfully!")
```

```python
# Right method

for i in range(3):
    if i == 2:
        continue
    print(i)
else:
    print("Loop finished successfully!")
```

## While Loop

```python
# Basic while Loop

i = 1
while i <= 5:
    print(i)
    i += 1

# 🧠 এখানে i <= 5 condition যতক্ষণ true, loop চলবে। প্রতি বার i এক করে বাড়ছে।
```

```python
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1
```

```python
# Decrementing (Reverse Loop)

i = 5
while i > 0:
    print(i)
    i -= 1

# 🧠 i প্রতি iteration-এ 1 করে কমছে।
```

```python
# Infinite Loop (⚠️ careful)

# while True:
#     print("This will run forever!")

# 🧠 যতক্ষণ পর্যন্ত manually stop না করা হয় (Ctrl + C), এই loop চলতে থাকবে।
# এ ধরনের loop আমরা সাধারণত game, menu বা background process এ ব্যবহার করি।
```

```python
# Sum of Numbers using while

i = 1
total = 0

while i <= 5:
    total += i
    i += 1

print("Sum =", total)

# 🧠 (1+2+3+4+5 = 15)
```

```python
# Even Numbers using while

i = 2
while i <= 10:
    print(i)
    i += 2
```

```python
# Reverse a Number using while

num = int(input("Enter a number: "))
rev = 0

while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed:", rev)
```

```python
# while with else

i = 1
while i <= 3:
    print(i)
    i += 1
else:
    print("Loop completed successfully!")

# 🧠 else অংশ তখনই execute হয় যখন loop naturally শেষ হয় (break ছাড়া)।

# Behavior:
# while loop চলবে যতক্ষণ i <= 3 true।
# যখন loop naturally শেষ হবে (condition false হবে), তখন else block execute হবে।
# যদি loop এর মধ্যে break হতো, তখন else execute হতো না।
```

```python
i = 1
while i <= 3:
    print(i)
    i += 1

print("Loop completed successfully!")
```

```python
i = 1
while i <= 3:
    print(i)
    if i == 2:
        break
    i += 1
else:
    print("Loop completed successfully!")
```

```python
i = 1
while i <= 3:
    print(i)
    if i == 2:
        break
    i += 1

print("Loop completed successfully!")
```

## Nested for loop

```python
# 🔷 What is a Nested Loop?
# যখন একটি loop-এর ভিতরে আরেকটি loop থাকে, সেটাকে বলা হয় nested loop।
# 👉 মানে, outer loop চলবে এবং প্রতিবার সেটির ভিতরে inner loop সম্পূর্ণভাবে চলবে।

for i in range(3):
    for j in range(3):
        print(i, "->", j, ".", "Hello Python")
```

```python
for i in range(3):
    for j in range(3):
        print(f"i = {i}, j = {j}")
```

```python
# Nested while loop

i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(f"i = {i}, j = {j}")
        j += 1
    i += 1
```

```python
# Simple Nested for loop

for i in range(1, 4):        # Outer loop
    for j in range(1, 4):    # Inner loop
        print(i, j)

# 🧠 Explanation:
# Outer loop: i = 1, 2, 3
# Inner loop: প্রতিবার i এর জন্য j = 1, 2, 3 চলে
```

```python
# Multiplication Table

for i in range(1, 4):
    for j in range(1, 6):
        print(f"{i} x {j} = {i*j}")
    print("-----")
```

```python
# Star Pattern (Common in Interviews ⭐)

for i in range(1, 6):
    for j in range(1, i + 1):
        print("*", end="")
    print()
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()
```

```python
# Number Triangle

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# Reverse Pattern

for i in range(5, 0, -1):
    for j in range(1, i + 1):
        print("*", end="")
    print()
```
