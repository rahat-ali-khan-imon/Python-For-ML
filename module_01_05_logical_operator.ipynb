## Logical Operators ---> and, or, not

```python
x, y, z = 10, 20, 30
a = 0
b = 1
c = -5

print(x and y)  # Output: 20 কারণ 'and' শেষ truthy value ফেরত দেয়
print(x or y)  # Output: 10 কারণ 'or' প্রথম truthy value ফেরত দেয়
print(not x)  #  not x -> not 10 -> not True -> False
print(not y)
print("\n")

print(not a)
print(not b)
print(not c)
```

```python
print(x > y and x > z)   # False and False → False
print(x > y or z > x)    # False or True → True
print(not (y > z))       # not Frue → True
print(x > y and not (y > z))  # False and not(False) → False and True → False
print(x < y and not (z < y))# True and not(False) = True and True = True
```

```python
# Basic Examples of and

x = 10
y = 20

print(x < y and y > 5)    # True and True → ✅ True
print(x > y and y > 5)    # False and True → ❌ False
print(x == 10 and y == 30)# True and False → ❌ False
```

```python
# Basic Examples of or

x = 10
y = 20

print(x < y or y < 10)    # True or False → ✅ True
print(x > y or y > 25)    # False or False → ❌ False
print(x == 10 or y == 30) # True or False → ✅ True
```

```python
# Basic Examples of not

x = 10
y = 20

print(not (x < y))    # not(True) → ❌ False
print(not (x > y))    # not(False) → ✅ True
print(not False)      # ✅ True
print(not True)       # ❌ False
```

```python
# Combine and, or, not

x = 10
y = 20
z = 30

print(x < y and z > y)         # True and True → ✅ True
print(x < y or z < y)          # True or False → ✅ True
print(not (x > y and y > z))   # not(False and False) → not(False) → ✅ True
print(x < y and not (z < y))   # True and not(False) → True and True → ✅ True
print(not (x < y or z > y))    # not(True or True) → not(True) → ❌ False
```

```python
# Truth Table Summary

# | Expression      | Result  |
# | --------------- | ------- |
# | True and True   | ✅ True  |
# | True and False  | ❌ False |
# | False and True  | ❌ False |
# | False and False | ❌ False |
# | True or True    | ✅ True  |
# | True or False   | ✅ True  |
# | False or True   | ✅ True  |
# | False or False  | ❌ False |
# | not True        | ❌ False |
# | not False       | ✅ True  |
```

```python
# Basic Examples

x = 10
y = 20
z = 10

print(x == y)   # 10 == 20 → False
print(x == z)   # 10 == 10 → True
print(x != y)   # 10 != 20 → True
print(x > y)    # 10 > 20 → False
print(x < y)    # 10 < 20 → True
print(y >= x)   # 20 >= 10 → True
print(x <= z)   # 10 <= 10 → True
```

```python
# Combined with Logical Operators --> তুলনা এবং লজিক্যাল অপারেটর একসাথে ব্যবহার করা যায়

x = 10
y = 20
z = 30

print(x < y and y < z)   # True and True → ✅ True
print(x < y or y > z)    # True or False → ✅ True
print(not (x > y))       # not(False) → ✅ True
```

```python
# Comparing Strings (Alphabetically)

# Python এ string comparison হয় alphabetical order (Unicode value) অনুযায়ী।

# 🧠 মনে রাখবে:
# ছোট হাতের অক্ষরের (a-z) Unicode মান বড় হাতের (A-Z) থেকে বড়।
# অর্থাৎ 'a' > 'A' ✅ True


print("apple" == "apple")   # ✅ True
print("apple" != "banana")  # ✅ True
print("apple" > "banana")   # ❌ False ('a' < 'b')
print("cat" < "dog")        # ✅ True
```

```python
# Comparing Multiple Values

x = 5
print(1 < x < 10)    # ✅ True (chained comparison)
print(5 < x < 10)    # ❌ False
print(1 < x <= 5)    # ✅ True

# 🧠 এইভাবে একাধিক comparison একসাথে লেখা যায়, যা mathematical inequality-এর মতো কাজ করে।
```

```python
# Comparing Lists or Tuples (Lexicographic Order)

print([1, 2, 3] == [1, 2, 3])   # ✅ True
print([1, 2, 3] != [1, 2, 4])   # ✅ True
print([1, 2, 3] < [1, 2, 4])    # ✅ True
print((1, 2) > (1, 1))          # ✅ True

# তুলনা হয় element-by-element, প্রথম পার্থক্য দেখেই সিদ্ধান্ত নেয়।
```

## Precedence and Associativity

```python
# Precedence (Priority) আর Associativity (Direction of execution)

eq = 10 + 10 / 2 - 5 * 2

eq
```

```python
# What is Precedence?
# Operator precedence মানে হলো — এক্সপ্রেশনে কোন অপারেটর আগে কাজ করবে (priority)।

x = 10 + 2 * 3
print(x)

# Output: 16 (কারণ * এর precedence + এর চেয়ে বেশি, তাই আগে 2 * 3 = 6, তারপর 10 + 6 = 16)
```

```python
# What is Associativity?
# Associativity বলে দেয় একই precedence-এর অপারেটর এক্সপ্রেশনে কোন দিক থেকে কাজ করবে — বাম দিক (Left-to-Right) না ডান দিক (Right-to-Left)।

print(10 - 5 + 2)

# Left-to-Right associativity → (10 - 5) + 2 = 7
```

```python
# Python Operator Precedence (High → Low)

# | Precedence | Operators                        | Associativity |
# | ---------- | -------------------------------- | ------------- |
# | Highest    | `()` (Parentheses)               | Left to Right |
# | 2          | `**` (Exponent)                  | Right to Left |
# | 3          | `*`, `/`, `//`, `%`              | Left to Right |
# | 4          | `+`, `-`                         | Left to Right |
# | 5          | `<`, `<=`, `>`, `>=`, `==`, `!=` | Left to Right |
# | 6          | `not`                            | Right to Left |
# | 7          | `and`                            | Left to Right |
# | 8          | `or`                             | Left to Right |
# | Lowest     | `=` (Assignment)                 | Right to Left |
```

```python
# Multiplication before Addition

x = 10 + 2 * 3
print(x)  # 16
```

```python
# Parentheses changes order

x = (10 + 2) * 3
print(x)  # 36
```

```python
# Exponent has right-to-left associativity

print(2 ** 3 ** 2)  # 2 ** (3 ** 2) = 2 ** 9 = 512
```

```python
# Division and Multiplication (same precedence, left-to-right)

print(100 / 5 * 2)
# (100 / 5) * 2 = 20 * 2 = 40
```

```python
# Logical Operators

x = 10
y = 20
z = 30

print(x < y and y < z)   # True and True → True
print(x < y or y > z)    # True or False → True
print(not x < y)         # not(True) → False
```

```python
# Combining Arithmetic + Comparison + Logical

x = 10
y = 20

print(not (x + 10 > y and y < 50))
# Step by step:
# (x + 10 > y) → (20 > 20) → False
# (y < 50) → True
# False and True → False
# not(False) → True ✅
```

```python
# Assignment is lowest precedence

x = 5
y = 10
x += y * 2   # x = x + (y * 2)
print(x)     # 5 + 20 = 25
```

```python
# Confusing but important

print(True or False and False)
# and has higher precedence → True or (False and False)
# → True or False → True ✅
```

```python
# 🎯 Trick to Remember Precedence Order:

# (P E M D A C N A O)

# Parentheses
# Exponent
# Multiplication
# Division
# Addition/Subtraction
# Comparison
# Not
# And
# Or
```
