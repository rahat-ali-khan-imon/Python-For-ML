## Single/Double/Triple Quotes

```python
# Double quotes outside, single quotes inside

print("It's a beautiful day!")

# Explanation:
# বাইরেরটা " "
# ভিতরে ' ব্যবহার করা হয়েছে, তাই কোনো সমস্যা হয়নি।
```

```python
# Single quotes outside, double quotes inside

print('He said, "Python is fun!"')

# Explanation:
# বাইরেরটা ' '
# ভিতরে " " দেওয়া হয়েছে, তাই error হবে না।
```

```python
# Both same quotes (❌ Error case)
# এই লাইনটা error দেবে, কারণ Python ভেবে নেয় 'It' শেষ হয়ে গেছে — 's a beautiful day!' আরেকটা আলাদা অংশ।

print('It's a beautiful day!') 
```

```python
# Use escape character (\)

print('It\'s a beautiful day!')

# Explanation:
# \ দিয়ে ' কে escape করা হয়েছে, তাই Python বোঝে এটা string-এর ভিতরের quote।
```

```python
# Both quotes together

print("He said, \"It's a sunny day!\"")
```

```python
print("He said, \"It's a sunny day!\"")
print('He said, "It's a sunny day!"')
```

```python
# Triple quotes for multiline (bonus)

print("""He said, "It's a good day!" """)
print("""He said, "It's a funny day!" """)

# Triple quotes (''' or """) ব্যবহার করলে single আর double দুইটিই একসাথে দেওয়া যায় — কোনো escape ছাড়াই।
```

```python
# Single/Double/Triple Quotes Same

a = 'Python'
b = "Python"
c = '''Python'''
d = """Python"""

print(a == b == c == d)

# মানে — 'Python', "Python", '''Python''', """Python""" সবগুলোর value একই (string data type), শুধু লেখার ধরন (syntax) আলাদা।
```

```python
# কেন দুই ধরণের quote লাগে?
# কারণ কখনও কখনও string এর ভিতরেও quote দরকার হয়।

text1 = "He said, 'Python is fun!'"
text2 = 'She replied, "Yes, it is!"'

print(text1)
print(text2)

# 🧠 এখানে:
# প্রথম লাইনে আমরা double quote ব্যবহার করেছি বাইরে, তাই ভিতরে single quote allowed।
# দ্বিতীয় লাইনে single quote বাইরে, তাই ভিতরে double quote allowed।
```

```python
# যদি string এর ভিতরে দুই রকম quote দরকার হয়?
# তাহলে escape character \ ব্যবহার করতে হয়

text = "He said, \"Python's syntax is easy.\""
print(text)
```

## Ternary/Conditional/Selection Operator

```python
# Normal if-else:

a = 10
b = 20

if a > b:
    max_value = a
else:
    max_value = b

print("Maximum =", max_value)
```

```python
# Same thing using Ternary Operator:

a = 10
b = 20
max_value = a if a > b else b
print("Maximum =", max_value)

# Syntax: <value_if_true> if <condition> else <value_if_false>

# 💡 Tips:
# Use nested ternary only for short logic.
# যদি condition বেশি complex হয় → normal if-elif-else ব্যবহার করো।
# Readability সবসময় মাথায় রাখো।
# এক লাইনে simple decision নেওয়ার জন্য nested ternary perfect।
```

```python
a, b = map(int, input().split())

max_val = a if a > b else b

print(max_val)
```

```python
# Bonus Tip: Short if–else (Ternary Operator)

n = int(input())

result = "Even" if n % 2 == 0 else "Odd"
print(result)

ans = f"{n} is even" if n % 2 == 0 else f"{n} is odd"
print(ans)
```

## Nested Ternary

```python
# Basic Syntax of Nested Ternary ---> Positive / Negative / Zero (Nested Ternary)

num = int(input("Enter a number: "))

result = "Positive" if num > 0 else "Negative" if num < 0 else "Zero"  # 🧠 এখানে nested ternary ব্যবহার হয়েছে (একটার ভিতরে আরেকটা)।
print(result)

# 🧠 Rule:
# 👉 if এবং else এর ভিতরেও আরেকটা ternary রাখা যায়।
# Python এটাকে right-to-left direction এ evaluate করে।
```

```python
# Find Largest of Three Numbers

a, b, c = 10, 25, 20
largest = a if (a > b and a > c) else b if (b > c) else c
print("Largest =", largest)
```

```python
# Odd/Even + Range Check

num = int(input("Enter a number: "))

result = "Even and > 10" if num % 2 == 0 and num > 10 else \
          "Even and ≤ 10" if num % 2 == 0 else \
          "Odd and > 10" if num > 10 else "Odd and ≤ 10"

print(result)

# backslash \ এর পরে space থাকা যাবে না (Python এটাকে continuation হিসেবে ধরবে না)
```

```python
num = int(input("Enter a number: "))

result = (
    "Even and > 10" if num % 2 == 0 and num > 10 else
    "Even and <= 10" if num % 2 == 0 else
    "Odd and > 10" if num > 10 else
    "Odd and <= 10"
)

print(result)

# যদি তোমার line লম্বা লাগে, তবে backslash না দিয়ে ( ) bracket দিয়ে লিখলে কোডটা clean দেখায়
```
