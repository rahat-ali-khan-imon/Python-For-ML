# Taking Input in Python

## User Input

```python
name = input("Enter any string: ")

print(name)
print(type(name))
```

```python
age = input("Enter any age: ")

print(age)
print(type(age))
```

```python
age = 23
print(age)
print(type(age))


name = "Hi, Hello, Python"
print(name)
print(type(name))
```

```python
num = input("Enter any number: ")   # string

print(num, type(num))


num = int(num)  # type casting

print(num, type(num))
```

```python
num = int(input("Enter any integer number: "))

print(num, type(num))
```

```python
height = input("Enter any height: ")

height = height + 0.2     # TypeError: can only concatenate str (not "float") to str

print(height, type(height))
```

```python
height = input("Enter any integer height: ")

height += "20"

print(height, type(height))


height = int(height)
height += 10

print(height, type(height))
```

```python
height = float(input("Enter any float number: "))

height = height + 0.2

print(height)
print(type(height))
print(f"{height}")
print(f"{height:.7}")
print(f"{height:.2f}")
```
